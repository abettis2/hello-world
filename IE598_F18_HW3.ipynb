## Listing 2-4: Quantile-Quantile Plot

```python
from scipy import stats
import matplotlib.pyplot as plt

stats.probplot(df["n_trades"], dist="norm", plot=plt)
plt.title("Q-Q plot")
plt.ylabel("n_trades")
plt.show()

stats.probplot(df["LIQ SCORE"], dist="norm", plot=plt)
plt.title("Q-Q plot")
plt.ylabel("Liquidity Score")
plt.show()
```

## Listing 2-6: Parallel Coordinates graph for real atrribute visualization 

```python
from pandas.plotting import parallel_coordinates

plt.figure
parallel_coordinates(
    df[['Bloomberg Composite Rating', 'n_trades', 'volume_trades', 'n_days_trade' , 'LIQ SCORE']], 
    'Bloomberg Composite Rating') 

plt.legend(loc = 'best',bbox_to_anchor=(0.5, -0.2),ncol = 2, fancybox = True, shadow = True)

plt.show()
```

## Listing 2-7 Cross Plotting Pairs of attributes

```python
plt.scatter(df["total_mean_size"], df["LIQ SCORE"])
plt.xlabel('total_mean_size')
plt.ylabel('LIQ SCORE')
plt.show()

plt.scatter(df["percent_intra_dealer"], df["n_trades"])
plt.xlabel('percent_intra_dealer')
plt.ylabel('n_trades')
plt.show()
```
